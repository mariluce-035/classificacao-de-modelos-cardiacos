```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
file_path_final_full_dataset = "../Database/df_database_dados_normalizados_aumentados_minmax.csv"
file_path_new_dataset = "../Database/df_database_dados_normalizados_aumentados_minmax_morfologicos.csv"
```

```python
df_final_full_dataset = pd.read_csv(file_path_final_full_dataset)
df_final_full_dataset.head()
```

```python
df_new_dataset = df_final_full_dataset.drop('idade',axis=1)
df_new_dataset.head()
```

```python
df_new_dataset = df_new_dataset.drop('genero',axis=1)
df_new_dataset.head()
```

```python
df_new_dataset = df_new_dataset.reset_index(drop=True)
df_new_dataset.head()
```

```python
df_new_dataset.to_csv(file_path_new_dataset, index = False)
```
