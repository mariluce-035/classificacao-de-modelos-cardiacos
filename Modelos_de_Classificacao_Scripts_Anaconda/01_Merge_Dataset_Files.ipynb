```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
df_classe = pd.read_csv("../Database/id.classe.csv")
df_classe.head()
```

```python
df_edl = pd.read_csv("../Database/id.edltotal.csv")
df_edl.head()
```

```python
df_genero_idade = pd.read_csv("../Database/id.genero.idade.csv")
df_genero_idade.head()
```

```python
df_hough = pd.read_csv("../Database/id.hough90total.csv")
df_hough.head()
```

```python
df_hough = df_hough.drop('Unnamed: 9',axis=1)
df_hough.head()
```

```python
df_spharms = pd.read_csv("../Database/id.spharmsoctl10.csv")
df_spharms.head()
```

```python
df_spharms = df_spharms.drop('Unnamed: 9',axis=1)
df_spharms.head()
```

```python
df_volume_massa = pd.read_csv("../Database/id.volume.massa.csv")
df_volume_massa.head()
```

```python
df_merge_1 = df_genero_idade.merge(df_volume_massa,on=["id"])
df_merge_1.head()
```

```python
df_merge_2 = df_merge_1.merge(df_edl,on=["id"])
df_merge_2.head()
```

```python
df_merge_3 = df_merge_2.merge(df_hough,on=["id"])
df_merge_3.head()
```

```python
df_merge_4 = df_merge_3.merge(df_spharms,on=["id"])
df_merge_4.head()
```

```python
df_merge_5 = df_merge_4.merge(df_classe,on=["id"])
df_merge_5.head()
```

```python
df_merge_5['target'] = -1
df_merge_5.head()
```

```python
for ind, row in df_merge_5.iterrows():
    if (df_merge_5.loc[ind,'classe']=='SAN'): df_merge_5.loc[ind,'target'] = 0
    if (df_merge_5.loc[ind,'classe']=='CMD'): df_merge_5.loc[ind,'target'] = 1
    if (df_merge_5.loc[ind,'classe']=='CMH'): df_merge_5.loc[ind,'target'] = 2

df_merge_5.head()
```

```python
#dados_vazios = (df_merge_5.isnull().sum()/len(df_merge_5['id']))*100
dados_vazios = df_merge_5.isnull().sum()
dados_vazios.head(5)
```

```python
# Verifica a quantidade de amostras para cada classe
df_merge_5['classe'].value_counts()
```

```python
# Verifica a quantidade de amostras para cada classe
df_merge_5['target'].value_counts()
```

```python
for ind, row in df_merge_5.iterrows():
    if (df_merge_5.loc[ind,'genero']=='F'): df_merge_5.loc[ind,'genero'] = 0
    if (df_merge_5.loc[ind,'genero']=='M'): df_merge_5.loc[ind,'genero'] = 1

df_merge_5.head()
```

```python
df_merge_5 = df_merge_5.drop('id',axis=1)
df_merge_5 = df_merge_5.drop('classe',axis=1)
df_merge_5.head()
```

```python
df_merge_5.to_csv("../Database/df_database_dados_sem_tratamento.csv", index = False)
```
