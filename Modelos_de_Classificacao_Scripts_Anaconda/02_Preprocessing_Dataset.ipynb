```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
df_raw_data = pd.read_csv("../Database/df_database_dados_sem_tratamento.csv")
df_raw_data.head()
```

```python
df_raw_data.isnull().sum().sort_values(ascending=False)[:]
```

```python
volume_mediana = df_raw_data['volume'].median()
massa_mediana = df_raw_data['massa'].median()
idade_mediana = df_raw_data['idade'].median()

print(volume_mediana)
print(massa_mediana)
print(idade_mediana)
```

```python
df_raw_data['volume'].fillna(volume_mediana, inplace=True)
df_raw_data['massa'].fillna(massa_mediana, inplace=True)
df_raw_data['idade'].fillna(idade_mediana, inplace=True)

df_raw_data.isnull().sum().sort_values(ascending=False)[:]
```

```python
df_raw_data.head(400)
```

```python
dataframe_indexes = df_raw_data.columns[:]
print(dataframe_indexes)
```

```python
features_indexes = df_raw_data.columns[0:-1]
print(features_indexes)
```

```python
X = df_raw_data.loc[:, features_indexes].values
X
```

```python
Y = df_raw_data.loc[:,['target']].values
Y
```

```python
#X = StandardScaler().fit_transform(X)
X = MinMaxScaler().fit_transform(X)
X
```

```python
df_normalized_data_minmax = pd.DataFrame(X, columns = features_indexes)
df_normalized_data_minmax.head()
```

```python
df_normalized_data_minmax['target'] = Y
df_normalized_data_minmax.head(400)
```

```python
df_normalized_data_minmax.to_csv("../Database/df_database_dados_normalizados_minmax.csv", index = False)
```
