```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
df_norm_data = pd.read_csv("../Database/df_database_dados_normalizados_minmax.csv")
df_norm_data.head()
```

```python
features_indexes = df_norm_data.columns[0:-1]
print(features_indexes)
```

```python
df_norm_data_cmh = df_norm_data[df_norm_data['target']==2]
df_norm_data_cmh
```

```python
X_CMH = df_norm_data_cmh.loc[:, features_indexes].values

X_CMH
```

```python
distances_cmh = np.array([[-1,-1,-1]])

for i in range(183):
    for j in range(i+1,183):
            #print(str(i)+' '+str(j))
            sum_sq = np.sum(np.square(X_CMH[i] - X_CMH[j]))
            distance = np.sqrt(sum_sq)
            distances_cmh = np.append(distances_cmh, [[i,j,distance]], axis=0)

distances_cmh
```

```python
distances_cmh = distances_cmh[distances_cmh[:,2].argsort()]

distances_cmh
```

# Parei aqui - Calcular as novas amostras artificiais para chegar em 200 instancias

```python
X_CMH_AUGM = X_CMH.copy()

X_CMH_AUGM
```

```python
const_lambda = 0.5

for i in range(1,18):
    #print(str(i))
    #print(distances_cmh[i])
    #print(int(distances_cmh[i][0]))
    #print(int(distances_cmh[i][1]))
    nova_amostra = ((X_CMH[int(distances_cmh[i][0])] - X_CMH[int(distances_cmh[i][1])]) * const_lambda) + X_CMH[int(distances_cmh[i][1])]
    #print(nova_amostra)
    X_CMH_AUGM = np.append(X_CMH_AUGM, [nova_amostra], axis=0)
```

```python
df_norm_augm_data_cmh = pd.DataFrame(X_CMH_AUGM, columns = features_indexes)
df_norm_augm_data_cmh.head(500)
```
