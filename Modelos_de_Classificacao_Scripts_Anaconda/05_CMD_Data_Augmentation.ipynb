```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
df_norm_data = pd.read_csv("../Database/df_database_dados_normalizados_minmax.csv")
df_norm_data.head()
```

```python
features_indexes = df_norm_data.columns[0:-1]
print(features_indexes)
```

```python
df_norm_data_cmd = df_norm_data[df_norm_data['target']==1]
df_norm_data_cmd
```

```python
X_CMD = df_norm_data_cmd.loc[:, features_indexes].values

X_CMD
```

```python
distances_cmd = np.array([[-1,-1,-1]])

for i in range(116):
    for j in range(i+1,116):
            #print(str(i)+' '+str(j))
            sum_sq = np.sum(np.square(X_CMD[i] - X_CMD[j]))
            distance = np.sqrt(sum_sq)
            distances_cmd = np.append(distances_cmd, [[i,j,distance]], axis=0)

distances_cmd
```

```python
distances_cmd = distances_cmd[distances_cmd[:,2].argsort()]

distances_cmd
```

# Parei aqui - Calcular as novas amostras artificiais para chegar em 200 instancias

```python
X_CMD_AUGM = X_CMD.copy()

X_CMD_AUGM
```

```python
const_lambda = 0.5

for i in range(1,85):
    #print(str(i))
    #print(distances_cmd[i])
    #print(int(distances_cmd[i][0]))
    #print(int(distances_cmd[i][1]))
    nova_amostra = ((X_CMD[int(distances_cmd[i][0])] - X_CMD[int(distances_cmd[i][1])]) * const_lambda) + X_CMD[int(distances_cmd[i][1])]
    #print(nova_amostra)
    X_CMD_AUGM = np.append(X_CMD_AUGM, [nova_amostra], axis=0)
```

```python
df_norm_augm_data_cmd = pd.DataFrame(X_CMD_AUGM, columns = features_indexes)
df_norm_augm_data_cmd.head(500)
```
