## Pipeline modelo

```python
def pipeline_model_run(test_name,classifier,random_seed,params,qtd_kfolds,refit_param,scoring):
    
#     #SMOTE
    pipeline = imbpipeline(steps = [['smote', SMOTE(random_state=random_seed)],
                               ['scaler', StandardScaler()],
                               ['classifier', classifier]])
    
    #NEARMIS
#     pipeline = imbpipeline(steps = [['undersampling', NearMiss()],
#                                ['scaler', StandardScaler()],
#                                ['classifier', classifier]])
    
    
    #RandomOverSampler
#     pipeline = imbpipeline(steps = [['RandomOverSampler', RandomOverSampler(random_state=random_seed)],
#                                ['scaler', StandardScaler()],
#                                ['classifier', classifier]])
    
    param_grid = params
    
    stratified_kfold = StratifiedKFold(n_splits=qtd_kfolds,
                                       shuffle=True,
                                       random_state=random_seed)
    
    
    grid_search = GridSearchCV(estimator=pipeline,
                           param_grid=param_grid,
                           scoring=scoring,
                           refit=refit_param,
                           cv=stratified_kfold,
                           n_jobs=-1,
                           return_train_score=True
                          )
    
    grid_search.fit(X_train, y_train)
    
    return grid_search
```

```python
def retorna_resultados(test_name,PASTA_RESULTADOS,grid_results,refit):
    
    result_all_params = pd.DataFrame(grid_results.cv_results_)[['params',
                                  'mean_test_accuracy',
                                  'mean_test_recall',
                                  'mean_test_precision',
                                  'mean_test_f1']]
    
    
    result_best_params = pd.DataFrame(grid_results.cv_results_)[['params',
                                  'mean_test_accuracy',
                                  'mean_test_recall',
                                  'mean_test_precision',
                                  'mean_test_f1']].query('mean_test_' + refit +   ' == mean_test_' + refit  +'.max()')
    
    
    acc=[]
    recall=[]
    precisao=[]
    f1=[]

    for i in range(0,10):
        acc.append(grid_results.cv_results_['split' +  str(i) + '_test_accuracy'][grid_results.best_index_])
        recall.append(grid_results.cv_results_['split' +  str(i) + '_test_recall'][grid_results.best_index_])
        precisao.append(grid_results.cv_results_['split' +  str(i) + '_test_precision'][grid_results.best_index_])
        f1.append(grid_results.cv_results_['split' +  str(i) + '_test_f1'][grid_results.best_index_])



    resultados = {'Acc': acc, 'Recall': recall, 'Precision': precisao, 'F1': f1}
    result_best_params_folds = pd.DataFrame(resultados)
    

    #resul_todos_param
    pd.DataFrame(result_all_params).to_excel(PASTA_RESULTADOS + '/' + NOME_TESTE + '_TODOS_PARAMETROS_MEDIA.xlsx',index=False)

    #result_melhor_param
    pd.DataFrame(result_best_params).to_excel(PASTA_RESULTADOS + '/' + NOME_TESTE + '_MELHOR_PARAMETRO_MEDIA.xlsx',index=False)

    #result_melhor_param_folds
    pd.DataFrame(result_best_params_folds).to_excel(PASTA_RESULTADOS + '/' + NOME_TESTE + '_MELHOR_PARAMETRO_FOLDS.xlsx')


    #SALVA MODELO    
    joblib.dump(grid_results.best_estimator_,  PASTA_RESULTADOS + '/'+ test_name + '.pkl')
    
    
    return result_all_params,result_best_params,result_best_params_folds
```

```python
def roda_teste(df_teste,modelo,nome_teste):
    X_teste = df_teste.drop(['TARGET'],axis=1)
    Y_teste = df_teste['TARGET']

    # scaler = StandardScaler().fit(X_teste)
    # X_teste = scaler.transform(X_teste)

    Ypred = modelo.predict(X_teste)




    #MOSTRA RESULTADOS
    metricas=[]
    metricas.append({ 'Tipo_teste':nome_teste,
                      'Accuracia':round(accuracy_score(y_true=Y_teste, y_pred=Ypred)*100,2),
                      'Recall':round(recall_score(y_true=Y_teste, y_pred=Ypred)*100,2),
                     'Precisao':round(precision_score(y_true=Y_teste, y_pred=Ypred)*100,2),
                      'F1':round(f1_score(y_true=Y_teste, y_pred=Ypred)*100,2)})

    resultados_test = pd.DataFrame(metricas,columns=['Tipo_teste','Accuracia','Recall','Precisao','F1'])

    return resultados_test
```
